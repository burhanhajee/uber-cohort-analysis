```python
print("="*50)
print("Checking activity_logs: session_end vs. session_start...")

# Find all rows where the end time is on or before the start time
invalid_sessions = activity_logs[activity_logs['session_end'] <= activity_logs['session_start']]

if len(invalid_sessions) == 0:
    print("PASS: All sessions have an end time after their start time.")
else:
    print(f"FAIL: Found {len(invalid_sessions)} sessions where session_end is NOT after session_start.")
    print("Here are some examples:")
    display(invalid_sessions.head())

print("="*50)
print("\n")
```

```python
print("="*50)
print("Checking trip_logs: pickup_time vs. request_time...")

# Find all rows where the pickup time is on or before the request time
invalid_trips = trip_logs[trip_logs['pickup_time'] <= trip_logs['request_time']]

if len(invalid_trips) == 0:
    print("PASS: All trips have a pickup time after their request time.")
else:
    print(f"FAIL: Found {len(invalid_trips)} trips where pickup_time is NOT after request_time.")
    print("Here are some examples:")
    display(invalid_trips.head())

print("="*50)
```
